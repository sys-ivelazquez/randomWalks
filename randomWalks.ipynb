# Modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Activity 1: Correlated Random Walk - 1 Trajectory

```python
# Number of steps
n_steps = 7500

# Correlation between steps
correlation = 0.9

# Generating a random walk
steps = np.random.randn(n_steps)
for i in range(1, n_steps):
    steps[i] += correlation * steps[i - 1]

# Generating the trajectory
trajectory = np.cumsum(steps)

# Plotting the trajectory
plt.plot(trajectory)
plt.title('Activity 1: Correlated Random Walk - 1 Trajectory')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.show()
```

# Activity 2: Lévy distribution - N different curves

```python
# Number of curves
n_curves = 5

# Lévy index
index = 1.5

# Generating Lévy distribution
plt.figure(figsize = (10, 6))
for _ in range(n_curves):
    levy_steps = np.random.standard_t(df = index, size = 1000)
    plt.plot(np.cumsum(levy_steps), label = f'Curve {_ + 1}')

# Plotting results
plt.title('Activity 2: Lévy distribution - N different curves')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.legend()
plt.show()
```
